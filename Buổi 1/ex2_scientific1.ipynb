```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display # hiển thị rich output
```

```python
%time
# generate randomly the center (x,y) of circle, color, area
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N)) ** 2 # 0 to 15 point radii
```

```python
# Hiển thị đồ thị
# %matplotlib inline # Dòng này chỉ cần thiết trong một số môi trường, nếu cần
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title("Biểu đồ phân tán ngẫu nhiên")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.show()
```
