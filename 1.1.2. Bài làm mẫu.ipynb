# NHIỆM VỤ 1: SỬ DỤNG DỮ LIỆU COVID-19

```python

# pip install pandas numpy scipy

import pandas as pd
import numpy as np
from scipy import stats

```

```python
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df = pd.read_csv(url)

# Xem cấu trúc dữ liệu
print(df.shape)
print(df.columns)
df.head()

```

```python
# Lọc bỏ các giá trị NaN
data = df['total_cases'].dropna()

print("Số lượng quan sát:", len(data))
print("5 giá trị đầu:\n", data.head())


```

```python
# Mean (trung bình)
mean_val = np.mean(data)

# Median (trung vị)
median_val = np.median(data)

# Mode (giá trị xuất hiện nhiều nhất)
mode_val = stats.mode(data, keepdims=True)[0][0]

# Variance (phương sai)
variance_val = np.var(data)

# Standard deviation (độ lệch chuẩn)
std_val = np.std(data)

# Range (khoảng)
range_val = np.max(data) - np.min(data)

# Percentile (ví dụ: 25th, 50th, 75th)
percentiles = np.percentile(data, [25, 50, 75])

# Quartiles
Q1, Q2, Q3 = np.percentile(data, [25, 50, 75])

# IQR (interquartile range)
IQR = Q3 - Q1

# In kết quả
print("Thống kê dữ liệu COVID:")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Mode: {mode_val:.2f}")
print(f"Variance: {variance_val:.2f}")
print(f"Standard Deviation: {std_val:.2f}")
print(f"Range: {range_val:.2f}")
print(f"Percentiles (25th, 50th, 75th): {percentiles}")
print(f"IQR: {IQR:.2f}")


```

```python
vn = df[df['location'] == 'Vietnam']['total_cases'].dropna()

print("Số lượng dữ liệu Việt Nam:", len(vn))
print("Mean:", np.mean(vn))
print("Median:", np.median(vn))
print("Std:", np.std(vn))
print("IQR:", np.percentile(vn, 75) - np.percentile(vn, 25))
```
